```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

#CARICAMENTO DEL CSV
file_path = r"C:\Users\Utente\Downloads\california_housing_data.csv"
df = pd.read_csv(file_path)

df.head()
```

```python
plt.style.use('_mpl-gallery')

np.random.seed(3)
x = 4 + np.random.normal(0, 2, 24)
y = 4 + np.random.normal(0, 2, len(x))

sizes = np.random.uniform(15, 80, len(x))
colors = np.random.uniform(15, 80, len(x))

# plot
fig, ax = plt.subplots()
ax.scatter(x, y, s=sizes, c=colors, vmin=0, vmax=100)
ax.set(
    xlim=(0, 8), xticks=np.arange(1, 8),
    ylim=(0, 8), yticks=np.arange(1, 8)
)

plt.show()
```

```python

plt.style.use('_mpl-gallery')

x = df["Longitude"].values
y = df["MedHouseVal"].values

# Campione ridotto per x2 e y2
sample = df.sample(25, random_state=3)
x2 = sample["Longitude"].values
y2 = sample["MedHouseVal"].values

# plot
fig, ax = plt.subplots()

ax.plot(x2, y2 + 25000, 'x', markeredgewidth=2)
ax.plot(x, y, linewidth=2.0)
ax.plot(x2, y2 - 25000, 'o-', linewidth=2)

ax.set(
    xlabel="Longitude",
    ylabel="Median House Value"
)

plt.show()
```
