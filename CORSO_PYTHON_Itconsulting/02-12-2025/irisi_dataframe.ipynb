```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_iris

# Load dataset
data = load_iris()  #load_iris: carica il dataset Iris, un dataset classico di esempio con 4 feature.
df = pd.DataFrame(data.data, columns=data.feature_names)  
#data.data contiene le feature (lunghezza e larghezza dei sepali e dei petali).
#data.feature_names sono i nomi delle colonne.
#data.target  la classe della pianta (0, 1, 2).
#Il DataFrame df quindi ha 4 colonne di feature + una colonna target.

df['target'] = data.target

# PCA a 2 dimensioni
pca = PCA(n_components=2)
components = pca.fit_transform(df[data.feature_names])  #array numpy con due colonne

df['PC1'] = components[:, 0]
df['PC2'] = components[:, 1]

plt.figure(figsize=(8,6))
plt.scatter(df['PC1'], df['PC2'], c=df['target'], alpha=0.8)
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('PCA 2D - Iris Dataset')
plt.grid(True)
plt.show()
```
